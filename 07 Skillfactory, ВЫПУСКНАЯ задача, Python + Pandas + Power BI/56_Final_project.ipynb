```python
from collections import Counter

# Создаём рекурсивную функцию, которая будет строить всевозможные пары из предоставленного списка номеров курсов
def user_doubles(my_list):
    # Создаём пустой объект Counter, который будет формировать в данной функции и потом возвращать
    doubles_proc = Counter()
    
    # Признак выхода из рекурсии - на вход функции было подано всего ДВА номера курса
    if len(my_list) == 2:
        # Формируем кортеж из двух курсов купленных одним студентом.
        # Именно по этому кортежу мы и будем проводить подсчёт пар, а потом и работать далее
        doubles_proc = Counter([(my_list[0], my_list[1])])
        return doubles_proc
    
    # Добавляем в Counter всевозможные пары, в которых участвует ПЕРВЫЙ элемент из высланного списка
    for my_user in range(len(my_list)-1):
         doubles_proc += Counter([(my_list[0], my_list[my_user + 1])])
    
    # Рекурсивно вызываем функцию дальше - передаём тот же список курсов, но БЕЗ ПЕРВОГО элемента
    doubles_proc += user_doubles(my_list[1:])
    
    #Возвращаем результат работы функции с построенными всеми парами курсов
    return doubles_proc
    

# Создаём пустой объект Counter, в котором будет хранить все нужные нам пары курсов
doubles_total = Counter()    
# Перебираем всех студентов и для каждого из них вызываем рекурсивную функцию
# Напомню, что список prom_list я создал чуть выше для проверки перекачанных данных из Metabase и он содержит всех нужных нам студентов
for user_list in prom_list:
    doubles_total += user_doubles(list(good_carts[good_carts['user_id'] == user_list]['resource_id'].values))

# Здесь я на экран вывошу получившиеся пары просто для того, чтобы проверить что результат получен и он похож на требуемый    
doubles_total.keys()
```

## 8. Ранжирование всех курсов по количеству пар курсов, в которых он участвует

2. В построненных нами парах НЕТ пар с двумя одинаковыми курсами.  

Поэтому нам нужно просто для каждого курса подсчитать количества пар, в которых он присутствует. А потом по полученному количеству провести ранжирование.  

### 8.1 Используем объект Counter

```python
# Вспомним какие данные объекта Counter у нас есть
doubles_total
```

```python
# Создадим функцию, которая для каждого номера курса будет определять количество вхождений в пары с помощью Counter
def getNumberOfCourseInDoubles(course_number):
    # Устанавливаем счётчик в 0
    prom_counter = 0
    # Суммируем в цикле по всем записям в Counter
    for prom_double in doubles_total:
        # Проверям входит ли высланный нам курс в кортеж из 2 курсов
        if prom_double[0] == course_number or prom_double[1] == course_number:
            # Если да, то добавляем значение Counter для данного элемента
            prom_counter += doubles_total[prom_double]
            
    return prom_counter

courses_rank_SQL['course_count_in_doubles'] = courses_rank_SQL['resource_id'].apply(getNumberOfCourseInDoubles)
courses_rank_SQL
            
```

```python
# Теоритически мы можем ещё разз упорядочить наш DataFrame с помощью sort_values и потом просто переписать в нужный нам слолбец индексы строк
# Но мы пойдём другим путём и продемонстрируем что умеем ранжировать и иначе :-)
def ranking_by_doubles(course_count): 
    return courses_rank_SQL[courses_rank_SQL['course_count_in_doubles'] >= course_count]['resource_id'].count()

# При просчёте суммарного ранжирования номера получаются не с 1 и не последовательны. Исправим это и сделаем правильное ранжирование
def renumbering_total_ranking(course_count2): 
    return courses_rank_SQL[courses_rank_SQL['course_rank_total'] <= course_count2]['resource_id'].count()

courses_rank_SQL['course_rank_by_doubles'] = courses_rank_SQL['course_count_in_doubles'].apply(ranking_by_doubles)
courses_rank_SQL['course_rank_total'] = courses_rank_SQL['course_rank_by_user'] + courses_rank_SQL['course_rank_by_doubles']
courses_rank_SQL.sort_values(by='course_rank_total', ascending=True, inplace=True)
courses_rank_SQL['course_rank_total'] = courses_rank_SQL['course_rank_total'].apply(renumbering_total_ranking)
courses_rank_SQL.reset_index(inplace=True, drop=True)
courses_rank_SQL
```

## 9. Наконец то мы добрались до расчёта столбцов с двумя рекомендуемыми курсами

### 9.1 Используем объект Counter

Будем использовать следующий алгоритм для выбора ДВУХ рекомендуемых курсов для каждого курса:  
1. Вначале из всех элементов объекта Counter отберём только те пары, в которых имеется интересующий нас курс.  
2. Из данных пар отбираем две с наибольшим количеством повторов.  
3. Из отобранных двух пар выбираем в качестве рекомендуемых вторые курсы, то есть пары к нашему курсу.  
4. Если не удастся найти ни одной пары к нашему курсу или только одну пару, то вставляем нули в рекомендуемые курсы. Что делать с нулями - мы решим чуть попозже.

```python
# Создаём пустые списки для обоих рекомендаций и количества пар рекомендуемых курсов с искомым
prom_list_course_1 = []
prom_list_count_1 = []
prom_list_course_2 = []
prom_list_count_2 = []

# Запускаем цикл по всем курсам
for prom_course in range(courses_rank_SQL.shape[0]):
    prom_Counter = Counter() # пустой Counter, в котором будем собирать все пары с рассчётным курсом
    prom_resourse = courses_rank_SQL.iloc[prom_course]['resource_id'] # фиксируем расчётный курс в отдельной переменной
    # устраиваем цикл по всем парам курсов
    for prom_double in doubles_total.elements():
        # Проверям входит ли расчётный курс в проверяемый
        if prom_double[0] == prom_resourse or prom_double[1] == prom_resourse:
            # Если да, то добавляем данную пару в промежуточный Counter
            prom_Counter += Counter([prom_double])
    
    # делаем проверку на то, что существует хотя бы ОДНА пара с рассчётным курсом
    if len(prom_Counter) > 0:
        # если да, то добавляем 1 рекомендованный курс в отдельный список
        if prom_Counter.most_common()[0][0][0] == prom_resourse:
            prom_list_course_1.append(prom_Counter.most_common()[0][0][1])
        else:
            prom_list_course_1.append(prom_Counter.most_common()[0][0][0])
        # отдельно храним количество пар с расчётным курсом
        prom_list_count_1.append(prom_Counter.most_common()[0][1])
    
    # делаем проверку на то, что существует хотя бы ДВЕ пары с рассчётным курсом
    if len(prom_Counter) > 1:
        # если да, то добавляем 1 рекомендованный курс в отдельный список
        if prom_Counter.most_common()[1][0][0] == prom_resourse:
            prom_list_course_2.append(prom_Counter.most_common()[1][0][1])
        else:
            prom_list_course_2.append(prom_Counter.most_common()[1][0][0])
        
        # отдельно храним количество пар с расчётным курсом
        prom_list_count_2.append(prom_Counter.most_common()[1][1])
    

# формируем в нашем датасете нужные колонки
courses_rank_SQL['course_recomend_1'] = pd.DataFrame(prom_list_course_1)
courses_rank_SQL['course_count_1'] = pd.DataFrame(prom_list_count_1)
courses_rank_SQL['course_recomend_2'] = pd.DataFrame(prom_list_course_2)
courses_rank_SQL['course_count_2'] = pd.DataFrame(prom_list_count_2)
courses_rank_SQL
```

Однако, у нас наблюдается ещё одна засада, с которой нам нужно разобраться.  
Что я имею в виду?  
Мы уже определились, что если какой то курс покупался со всеми другими не более level_for_leave_recomendation, то полученные рекомендации №1 и №2 мы заменим на самые высокорейтинговые с соответствии с ранжированием в колонке course_rank_total.  
Здесь заложен недостаток. Нам нужно всего 2 рекомендации для РАЗНЫХ курсов. Поэтому для разных курсов ВСЕГДА будут выдаваться два самых высорейтинговых курса. Очевидно, что это не есть хорошо.  
Правильно было бы, если бы выбрали некоторое количество высорейтинговых курсов и выдавали бы их в рекомендации по очереди и по кругу. 
Но для этого нам нужно зафиксировать ещё один параметр - количество данных курсов, которые будут подставляться в рекомендации по очереди и по кругу. И вновь, если Георгию в будущем НЕ понравится наше количество, то он всегда сможет ввести своё значение, перезапустит код на выполнение и получит нужные ему рекомендации.  
Так сделаем же это - введём ещё одну переменную...  

```python
# ВНИМАНИЕ! ВНИМАНИЕ! ВНИМАНИЕ!
# Сейчас будет задано ЗНАЧЕНИЕ переменной, которая определяет количество самых высорейтинговых курсов, которые будут заменять рекомендации курсам, 
# у которых низкое количество пересечений с другими курсами - не больше level_for_leave_recomendation раз.

number_of_most_ranking_for_recomendation = 20   # и вновь сам придумал эту цифру. Если кто-то не согласен с ней - замени на своё значение. :-)
```

```python
# Задаём переменную, которая будет по очереди перебирать рекомендации на замену из пула до number_of_most_ranking_for_recomendation
prom_ranking_for_recomendation = int(0)

# Запускаем цикл по всем курсам
for prom_course in range(courses_rank_SQL.shape[0]):
    # Проверяем стоит ли заменять первую рекомендацию
    if courses_rank_SQL.iloc[prom_course]['course_count_1'] <= level_for_leave_recomendation:
        # делаем замену ПЕРВОЙ рекомендации
        # в силу того, что все строки в датафреймы упорядочены и проиндексированы в соответствии с ранжированием course_rank_total,
        # то это нам существенно облегчает перебор рекомендуемых замен
        courses_rank_SQL.iloc[prom_course]['course_recomend_1'] = courses_rank_SQL.iloc[prom_ranking_for_recomendation]['resource_id']
        prom_ranking_for_recomendation += 1
        # делаем проверку на то, чтобы prom_ranking_for_recomendation не превосходил number_of_most_ranking_for_recomendation
        # если такое происходит - сбрасываем счётчик в 0, тем самым его перезапуская
        if prom_ranking_for_recomendation >= number_of_most_ranking_for_recomendation:
            prom_ranking_for_recomendation = 0
    
    if courses_rank_SQL.iloc[prom_course]['course_count_2'] <= level_for_leave_recomendation:
        # делаем замену ВТОРОЙ рекомендации
        courses_rank_SQL.iloc[prom_course]['course_recomend_2'] = courses_rank_SQL.iloc[prom_ranking_for_recomendation]['resource_id']
        prom_ranking_for_recomendation += 1
        # делаем проверку на то, чтобы prom_ranking_for_recomendation не превосходил number_of_most_ranking_for_recomendation
        # если такое происходит - сбрасываем счётчик в 0, тем самым его перезапуская
        if prom_ranking_for_recomendation >= number_of_most_ranking_for_recomendation:
            prom_ranking_for_recomendation = 0
            
courses_rank_SQL
```
